```python

#Pertemuan 12: Data Cleaning and Preparation using Pandas
# Exercise 1: Identifying and Handling Missing Data

import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Tika', 'Bob', 'Charlie', 'Musa', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
import pandas as pd

# Muat dataset
data = pd.read_csv("athlete_events.csv")

# Pilih kolom kategori (kolom bertipe 'object')
categorical_columns = data.select_dtypes(include=['object']).columns

# Standarisasi kolom kategori
for column in categorical_columns:
    # Ubah semua teks menjadi huruf kecil
    data[column] = data[column].str.lower()
    # Hilangkan spasi tambahan di awal dan akhir
    data[column] = data[column].str.strip()

# Tampilkan hasil standarisasi pada 5 baris pertama
print("Kolom kategori setelah standarisasi:\n")
print(data[categorical_columns].head())

# Simpan hasil yang sudah dimodifikasi ke file baru
data.to_csv("athlete_events_standardized.csv", index=False)
print("\nData telah disimpan ke file 'athlete_events_standardized.csv'")
```

```python
import pandas as pd

# Load dataset
file_path = 'athlete_events.csv'  # Lokasi file yang Anda unggah
data = pd.read_csv(file_path)

# Mengecek data awal
print("Data sebelum pembersihan:\n", data.head())
print("\nJumlah nilai hilang per kolom sebelum pembersihan:\n", data.isnull().sum())

# Mengisi nilai yang hilang
# Kolom numerik
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Height'] = data['Height'].fillna(data['Height'].median())
data['Weight'] = data['Weight'].fillna(data['Weight'].median())

# Kolom kategorikal
data['Sex'] = data['Sex'].fillna('Tidak Diketahui')
data['Medal'] = data['Medal'].fillna('Tidak Diketahui')

# Menghapus baris dengan nilai hilang pada kolom kritis
# Misalnya, kolom 'Team' dianggap sebagai kolom kritis
data = data.dropna(subset=['Team'])

# Mengecek hasil setelah pembersihan
print("\nSetelah pembersihan:\n", data.head())
print("\nJumlah nilai hilang per kolom setelah pembersihan:\n", data.isnull().sum())

# Menyimpan data yang telah diproses
output_path = 'processed_athlete_events.csv'
data.to_csv(output_path, index=False)
print(f"\nData telah diproses dan disimpan ke file: {output_path}")
```

```python
# Menghitung IQR untuk menangani outlier
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Batasi nilai outlier (capping)
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Identifikasi kolom numerik yang relevan
numerical_columns = ['Age', 'Height', 'Weight']  # Kolom numerik pada dataset

# Terapkan penanganan outlier untuk setiap kolom numerik
for col in numerical_columns:
    data = handle_outliers(data, col)

# Periksa hasil setelah penanganan outlier
print("Data setelah penanganan outlier:\n", data[numerical_columns].describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Mengelompokkan data berdasarkan 'City' dan 'Sex' untuk menghitung jumlah atlet per kota dan jenis kelamin
grouped_data = data.groupby(['City', 'Sex']).size().unstack(fill_value=0)

# Data untuk plot
categories = grouped_data.index
men_counts = grouped_data['M']  # Jumlah pria
women_counts = grouped_data['F']  # Jumlah wanita

# Plot
x = np.arange(len(categories))  # Posisi X untuk setiap kategori
width = 0.4  # Lebar bar, lebih besar untuk jarak antar bar

fig, ax = plt.subplots(figsize=(12, 7))

# Membuat dua bar chart untuk pria dan wanita
bar1 = ax.bar(x - width/2, men_counts, width, label='Male', color='b')
bar2 = ax.bar(x + width/2, women_counts, width, label='Female', color='r')

# Menambahkan label dan judul
ax.set_xlabel('City', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Jumlah Atlet Berdasarkan Gender dan Kota', fontsize=16, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right', fontsize=10)
ax.legend()

# Menambahkan jarak antar label dan sumbu
ax.tick_params(axis='x', pad=10)

# Menambahkan nilai di atas batang
for bar in bar1 + bar2:
    height = bar.get_height()
    ax.annotate(f'{int(height)}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom')

# Menambahkan grid horizontal untuk kemudahan pembacaan
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Memuat dataset
data = pd.read_csv('athlete_events.csv')

# Mengisi nilai kosong yang diperlukan untuk memastikan data lengkap
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Sex'] = data['Sex'].fillna('Unknown')
data['NOC'] = data['NOC'].fillna('Unknown')
data['Sport'] = data['Sport'].fillna('Unknown')

# Bar plot untuk jumlah atlet berdasarkan gender
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Sex', palette='coolwarm')  # Menggunakan palet 'coolwarm'
plt.title('Jumlah Atlet Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Jumlah Atlet')
plt.show()

# Scatter plot: Usia vs Tinggi badan, dikelompokkan berdasarkan gender
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Height', hue='Sex', palette='viridis')  # Menggunakan palet 'viridis'
plt.title('Scatter Plot: Usia vs Tinggi Badan')
plt.xlabel('Usia')
plt.ylabel('Tinggi Badan (cm)')
plt.legend(title='Gender')
plt.show()

# Pie chart untuk distribusi jumlah atlet berdasarkan negara (NOC)
noc_counts = data['NOC'].value_counts().head(10)  # Menampilkan 10 negara teratas
plt.figure(figsize=(7, 7))
plt.pie(
    noc_counts,
    labels=noc_counts.index,
    autopct='%1.1f%%',
    colors=sns.color_palette("cool", len(noc_counts))  # Menggunakan palet 'cool'
)
plt.title('Distribusi Atlet Berdasarkan Negara (Top 10)')
plt.show()

# Box plot: Distribusi usia berdasarkan olahraga
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=data.dropna(subset=['Age']),
    x='Sport',
    y='Age',
    palette='Spectral'  # Menggunakan palet 'Spectral'
)
plt.title('Box Plot: Distribusi Usia Berdasarkan Olahraga')
plt.xlabel('Olahraga')
plt.ylabel('Usia')
plt.xticks(rotation=45, ha='right')
plt.show()
```
